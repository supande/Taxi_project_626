```python
import csv
fn = 'trip_data_8.csv'
f = open(fn,'r')
reader = csv.reader(f)
n=0
shist = {}

for row in reader:
    if n>0:
        medallion= row[0]
    
        if len(medallion)!=0:
            if medallion in shist.keys():
                shist[medallion] += 1
            else:
                shist[medallion] = 1
        
    n+=1

shist.keys()

f.close()
```
